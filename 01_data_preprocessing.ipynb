```python
import pandas as pd

# Assuming df2 is already loaded
# Step 1: Numeric columns → fill NaN with mean
numeric_cols = df2.select_dtypes(include='number').columns
df2[numeric_cols] = df2[numeric_cols].fillna(df2[numeric_cols].mean())

# Step 2: Categorical columns → fill NaN with mode (most frequent value)
categorical_cols = df2.select_dtypes(include='object').columns
for col in categorical_cols:
    df2[col].fillna(df2[col].mode()[0], inplace=True)

# Step 3: Check missing values
print(df2.isnull().sum())
```
